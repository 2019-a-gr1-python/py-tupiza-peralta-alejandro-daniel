## Analisis a mercado libre de los siguientes productos: impresoras y otros aparatos electronicos

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from pandas.tools.plotting import table
```

```python
path_impresoras = 'C:/Users/ad_tp/OneDrive/Documentos/GitHub/py-tupiza-peralta-alejandro-daniel/Deberes/Proyecto/proyecto_scrapy/scrapyP/impresoras_items.csv'
```

```python
columnas_a_usar = ['titulo','modelo','marca',
                  'precio','condicion','opiniones',
                   'tipo_vendedor','ventas_vendedor']
```

```python
df = pd.read_csv(
        path_impresoras,
        usecols=columnas_a_usar,
        index_col='titulo'
        )
df_impresoras = pd.read_csv(
        path_impresoras,
        usecols=columnas_a_usar,
        )
pv_primeros_cinco= df_impresoras.filter(items = ["titulo","ventas_vendedor"]).sort_values('ventas_vendedor', ascending=False).head(5)
```

```python
plt.figure(figsize=(16,8))
# plot chart
ax1 = plt.subplot(121, aspect='equal')

pv_primeros_cinco.plot(kind='pie', y = 'ventas_vendedor', ax=ax1, autopct='%1.1f%%', 
 startangle=90, shadow=False, labels=pv_primeros_cinco['titulo'], fontsize=10)
plt.axis('equal')

# View the plot
plt.show()
```

```python
pv_primeros= df.filter(items = ["titulo","ventas_vendedor"]).sort_values('ventas_vendedor', ascending=False).head(5)

ax2 = plt.subplot(122)
plt.axis('off')
tbl = table(ax2, pv_primeros, loc='center')
tbl.auto_set_font_size(False)
tbl.set_fontsize(11)
plt.show()
```

# Las impresoras "Hp Deskjet 1115 Inyeccion Color F5s21a" y "Multifuncional Tinta Continua Brother Mfc-t910dw Duplex Wifi" tienen mas ventas

```python
pv_primeros_cinco.groupby('titulo').plot(kind='bar',stacked=True,title="Ventas segundo el id")
```

```python
marca_productos= df_impresoras.filter(items = ["titulo","marca"])

marca_productos.groupby('titulo').count()["marca"].plot(kind='bar',stacked=True,title="Total")
```

```python
mar= df.filter(items = ["titulo","marca"])
ax2 = plt.subplot(122)
plt.axis('off')
tbl = table(ax2, mar, loc='center')
tbl.auto_set_font_size(False)
tbl.set_fontsize(11)
plt.show()
```

```python
pv = df_impresoras.filter(items = ["titulo","ventas_vendedor"]).sort_values('ventas_vendedor', ascending=False).tail(12)
datos = pv
datos.plot(kind="bar",title="Ventas por id de productos vendidos")
```

```python
pv_tabla = df.filter(items = ["titulo","ventas_vendedor"]).sort_values('ventas_vendedor', ascending=False).tail(12)

ax2 = plt.subplot(122)
plt.axis('off')
tbl = table(ax2, pv_tabla, loc='center')
tbl.auto_set_font_size(False)
tbl.set_fontsize(11)
plt.show()
```

```python
titulos = np.unique(df_impresoras['titulo'].get_values(), return_counts = True)[0]
num_opiniones = df_impresoras['opiniones'].get_values()

posicion_y = np.arange(len(num_opiniones))
plt.figure(figsize=(5,15))
plt.barh(posicion_y, num_opiniones, align = "center")
plt.yticks(posicion_y, titulos)
plt.xlabel('Valoracion', fontsize=15)
plt.title("Cantidad de valoracion del vendedor por producto", fontsize=15)
for i, v in enumerate(num_opiniones):
    plt.text(v + 3, i, str(v),fontsize='x-large')
plt.show()
```

```python
titulos = np.unique(df_impresoras['titulo'].get_values(), return_counts = True)[0]
precios = df_impresoras['precio'].get_values()

posicion_y = np.arange(len(precios))
plt.figure(figsize=(10,15))
plt.barh(posicion_y, precios, align = "center")
plt.yticks(posicion_y, titulos)
plt.xlabel('Precio', fontsize=15)
plt.title("Precio de cada producto", fontsize=20)
for i, v in enumerate(precios):
    plt.text(v + 3, i, str(v),fontsize='x-large')
plt.show()
```
